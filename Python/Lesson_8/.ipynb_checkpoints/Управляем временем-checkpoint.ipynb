```python
bikes = pd.concat([pd.read_csv('bikes_q2_sample_apr.csv'),
                    pd.read_csv('bikes_q2_sample_may.csv'),
                    pd.read_csv('bikes_q2_sample_jun.csv'),
                    pd.read_csv('bikes_q3_sample_july.csv'),
                    pd.read_csv('bikes_q3_sample_aug.csv'),
                    pd.read_csv('bikes_q3_sample_sep.csv'),
                    pd.read_csv('bikes_q4_sample_oct.csv'),
                    pd.read_csv('bikes_q4_sample_nov.csv'),
                    pd.read_csv('bikes_q4_sample_dec.csv')])
```

```python
bikes_summer = bikes.loc['2018-06-01':'2018-08-31']
```

```python
pd.Series(bikes_summer.index).dt.day_name()
```

```python
bikes_summer.index = bikes_summer.index.day_name()
bikes_summer.index.name = 'Day'
bikes_summer_new = bikes_summer.reset_index()
```

```python
agg = bikes_summer_new.groupby(['Day', 'to_station_name']) \
            .aggregate({'to_station_id': 'count'}) \
            .rename(columns={'to_station_id': 'count'})
```

```python
agg.loc['Saturday'].sort_values('count', ascending=False)
```
